1. What is the difference between a function and a method in Python?
Answer- A function is a block of code that performs a specific task and is defined independently using the def keyword. A method, on the other hand, is a function that belongs to an object and is called using dot notation. For example, len() is a function, while "text".upper() is a method.

2. Explain the concept of function arguments and parameters in Python.
Answer- Parameters are the variable names listed in a function’s definition. Arguments are the actual values passed to the function when it is called. Parameters receive the values supplied through arguments during the function call.

3. What are the different ways to define and call a function in Python?
Answer- Functions can be defined using the def keyword or as anonymous functions using lambda. They can be called by writing the function name followed by parentheses containing any required arguments. Functions can also be built-in, user-defined, or lambda (anonymous) functions.

4. What is the purpose of the return statement in a Python function?
Answer- The return statement is used to send the result of a function back to the place where the function was called. It ends the function’s execution and provides a value. Without a return statement, the function will return None by default.

5. What are iterators in Python and how do they differ from iterables?
Answer- An iterable is an object that can be looped over (like a list or string). An iterator is an object that produces elements one at a time using the next() method. Iterators are created from iterables using the iter() function. The main difference is that iterators maintain their state and can fetch elements one by one.

6. Explain the concept of generators in Python and how they are defined.
Answer- Generators are special types of functions that return values one at a time using the yield keyword instead of return. They automatically create an iterator and maintain their internal state between each yield. Generators are useful for handling large or infinite data streams.

7. What are the advantages of using generators over regular functions?
Answer- Generators are memory-efficient because they yield one item at a time instead of storing the entire result in memory. They use lazy evaluation, meaning they generate values only when needed. This makes them ideal for working with large datasets or infinite sequences.

8. What is a lambda function in Python and when is it typically used?
Answer- A lambda function is a small anonymous function created using the lambda keyword. It can take any number of arguments but only one expression. Lambda functions are typically used for short, simple operations, often as arguments to functions like map(), filter(), or sorted().

9. Explain the purpose and usage of the map() function in Python.
Answer- The map() function is used to apply a specific function to each item of an iterable (such as a list). It returns a new iterable with the modified items. This is useful when performing the same operation on all elements of a collection.

10. What is the difference between map(), reduce(), and filter() functions in Python?
Answer- The map() function applies a function to each item of an iterable and returns a new iterable with the results. The filter() function filters out elements based on a condition and returns only those that satisfy the condition. The reduce() function (from the functools module) applies a function cumulatively to the items of an iterable, reducing them to a single result, such as a total sum or product.
